```python
from google_play_scraper import Sort, reviews_all
import csv

app_id = 'in.redbus.android'
sort_type = Sort.NEWEST

reviews = reviews_all(
    app_id,
    lang='id',
    country='id',
    sort=sort_type
)

with open('ulasanRedbus.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile, delimiter=';', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    writer.writerow(['reviewId', 'review', 'date'])
    for r in reviews:
        writer.writerow([r['reviewId'], r['content'], r['at']])
```
